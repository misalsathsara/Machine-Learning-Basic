# **Selection with multiple Inputs**

```python
m=35
if m>=75:
  grade="A"
elif m>=65:
  grade="B"
elif m>=55:
  grade="C"
else:
  grade="F"
print(grade)
#nested if condition
```

```python
n=-4
if n>0:
  print("positive")
elif n<0:
  print("negative")
else:
  print("zero")
```

# **Working with random**

```python
import random #firts you have to import random module
print(random.random())
```

```python
print(random.uniform(5,10))
#random.uniform(a,b):returens a random float between a and b
```

```python
print(random.randint(5,10))
#random.randint(a,b):returns a random integer between a and b(inclusive).simulate a dice roll
```

```python
print(random.randrange(1,10,2))
#random.randrange(start,sstop,[,step]):
#returen a randomly selected element from rans (start,stop,step)
```

```python
colors=['red','blue','green']
print(random.choice(colors))
#random.choice(seq):returns a random element from a non-empty sequence
```

```python
deck=[1,2,3,4,5]
random.shuffle(deck)
#random.shuffles(x):shuffles the sequence in place
print(deck)
```

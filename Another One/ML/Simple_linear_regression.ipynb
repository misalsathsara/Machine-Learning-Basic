```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
x=np.array([800,1000,1300,1500,1750])
y=np.array([1.4,1.7,1.8,2.4,2.7])
```

```python
plt.scatter(x,y)
plt.xlabel('Sqaure Feet')
plt.ylabel('Cost')
plt.title('House Price Prediction')
plt.show()
```

```python
m,c=np.polyfit(x,y,1)
plt.xlabel('Sqare Feet')
plt.plot(x,y,'o')
plt.plot(x,m*x+c) #draw a line to evaluate the model using linear regression
```

```python
from sklearn.metrics import mean_squared_error
#going to use mean squared error
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([800,1000,1300,1500,1750]).reshape(-1,1)
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
r_squared = r2_score(y, y_pred)
print("R-squared:", r_squared)
```

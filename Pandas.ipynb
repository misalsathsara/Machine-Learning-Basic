```python
df5 = pd.DataFrame({
    'A': [1, 2, None, 4],
    'B': [None, 2, 7, 8],
    'C': [9, None, None, 12]
})
```

```python
df5
```

```python
df5.isnull()
```

```python
#count total null value in the entire dataframe
total_nulls = df5.isnull().sum().sum()
print(total_nulls)
```

```python
#count NaN values per column
nulls_per_column = df5.isnull().sum()
print(nulls_per_column)
```

```python
# Create a dictionary with the data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 38, 42]
}

# Create the DataFrame
df1 = pd.DataFrame(data)

# Display the DataFrame
print(df1)
```

**Filter and modify Data**

```python
filtered_df = df1[df1['Age'] > 30] # Select rows where Age is greater than 30
print("Rows where Age is greater than 30: ")
print(filtered_df)
```

```python
df1['Name'] = df1['Name'].str.upper() # Convert the 'Name' column to uppercase
df1['Age'] = df1['Age'] + 1 # Increase all ages by 1

print(df1)
```

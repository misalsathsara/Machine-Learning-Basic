```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, roc_auc_score, r2_score
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
```

```python
# allocate data to x and y
x = data.drop(['y'], axis=1)
y = data['y']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smt = SMOTE()
x, y = smt.fit_resample(x, y)
```

```python
y.value_counts()
```

```python
# split data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
# trainn the model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# get the predictions
y_pred = model.predict(x_test)
```

```python
# check the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ', accuracy)
```

```python
# classification report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
```

```python
cm1 = BaggingClassifier()
cm1.fit(x_train, y_train)
```

```python
y_pred1 = cm1.predict(x_test)
```

```python
# check the accuracy of the model
accuracy = accuracy_score(y_test, y_pred1)
print('Accuracy: ', accuracy)
```

```python
cm2 = BaggingRegressor()
cm2.fit(x_train, y_train)
```

```python
y_pred2 = cm2.predict(x_test)
```

```python
# check accuracy in BaggingRegressor model
accuracy = accuracy_score(y_test, y_pred2.round())
print('Accuracy: ', accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
```

```python
cm3 = RandomForestClassifier()
cm3.fit(x_train, y_train)
```

```python
y_pred3 = cm3.predict(x_test)
```

```python
# accuracy in RandomForestClassifier model
accuracy = accuracy_score(y_test, y_pred3)
print('Accuracy: ', accuracy)
```

```python
from sklearn.ensemble import ExtraTreesClassifier, ExtraTreesRegressor
```

```python
cm4 = ExtraTreesClassifier()
cm4.fit(x_train, y_train)
```

```python
y_pred4 = cm4.predict(x_test)
```

```python
# accuracy in ExtraTreesClassifier model
accuracy = accuracy_score(y_test, y_pred4)
print('Accuracy: ', accuracy)
```

**Regression**

```python
data = pd.DataFrame(pd.read_csv('datasets/Boston for LR.csv'))
```

```python
data.head()
```

```python
# allocate variables
x = data.drop(['medv'], axis=1)
y = data['medv']
```

```python
#split into training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
rgt = DecisionTreeRegressor(max_depth=3)
rgt.fit(x_train, y_train)
```

```python
#get the predictions
y_pred = rgt.predict(x_test)
```

```python
# get the accuracy
r_squared = r2_score(y_test, y_pred)
print('R Squared: ', r_squared)
```

```python
# get model score
rgt.score(x_test, y_test)
```

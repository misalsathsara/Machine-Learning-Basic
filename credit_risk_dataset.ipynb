```python
# Impute missing values with the median of the column
df['person_emp_length'].fillna(df['person_emp_length'].median(), inplace=True)

# Impute missing values for 'loan_int_rate' column
df['loan_int_rate'].fillna(df['loan_int_rate'].median(), inplace=True)
```

```python
df2 = df

# Specify the columns to impute outliers
columns_to_impute = ['person_age', 'person_income', 'person_emp_length', 'cb_person_cred_hist_length']

# Loop through each column and impute outliers
for column in columns_to_impute:
    # Calculate IQR
    Q1 = df2[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df2[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Impute outliers with the median
    median_value = df2[column].median()  # Calculate median
    df2[column] = df2[column].clip(lower=lower_bound, upper=upper_bound)  # Cap outliers
    df2.loc[(df2[column] < lower_bound) | (df2[column] > upper_bound), column] = median_value  # Replace with median

# Display the updated dataset
print("Updated Dataset after imputing outliers:")
df2.head()
```

```python
# Check the distribution of the target variable
class_distribution = df2_cleaned['loan_status'].value_counts()
print(class_distribution)

# Plot the class distribution
class_distribution.plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.show()

# Calculate the imbalance ratio
imbalance_ratio = class_distribution[0] / class_distribution[1]
print(f"Imbalance Ratio: {imbalance_ratio:.2f}")
```

```python
from sklearn.preprocessing import OneHotEncoder
from imblearn.over_sampling import SMOTE

# Separate features and target
X = df2_cleaned.drop('loan_status', axis=1)
y = df2_cleaned['loan_status']

# One-hot encode categorical columns
categorical_columns = ['person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file']
X_encoded = pd.get_dummies(X, columns=categorical_columns, drop_first=True)

# Apply SMOTE for oversampling
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_encoded, y)

# Check the new distribution
print("Resampled Class Distribution:")
print(y_resampled.value_counts())
```

```python
from sklearn.model_selection import train_test_split

# Split the resampled dataset
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Check the shapes of the splits
print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
model = RandomForestClassifier(random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
# Print the predicted class labels
print("Predicted Class Labels:")
print(y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# ROC-AUC score
print("ROC-AUC Score:")
print(roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))
```
